```python
import re
```

```python
def check_password_strength(password):
    strength = 0
    suggestions = []

    # Length check
    if len(password) >= 8:
        strength += 1
    else:
        suggestions.append("Use at least 8 characters")

    # Lowercase
    if re.search("[a-z]", password):
        strength += 1
    else:
        suggestions.append("Add lowercase letters (a-z)")

    # Uppercase
    if re.search("[A-Z]", password):
        strength += 1
    else:
        suggestions.append("Add uppercase letters (A-Z)")

    # Numbers
    if re.search("[0-9]", password):
        strength += 1
    else:
        suggestions.append("Include at least one number (0-9)")

    # Special characters
    if re.search("[@#$%^&*]", password):
        strength += 1
    else:
        suggestions.append("Add special characters (@#$%^&*)")

    # Strength result
    if strength <= 2:
        level = "Weak ❌"
    elif strength <= 4:
        level = "Medium ⚠️"
    else:
        level = "Strong ✅"

    return level, suggestions


# Input from user
password = input("Enter your password: ")

level, suggestions = check_password_strength(password)

print("\nPassword Strength:", level)

# Show suggestions if not strong
if suggestions:
    print("\nSuggestions to improve:")
    for s in suggestions:
        print("-", s)
else:
    print("\nGreat! Your password is strong 💪")
```

```python
def chatbot():
    print("🤖 Chatbot Started! Type 'exit' to stop.\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Bot: Goodbye! 👋")
            break

        # Basic responses
        elif "hello" in user_input.lower():
            print("Bot: Hi there! 👋")

        elif "how are you" in user_input.lower():
            print("Bot: I'm just a program, but I'm doing great! 😄")

        elif "sports" in user_input.lower():
            print("Bot: I love sports! Which one do you like? ⚽🏏🏀")

        elif "name" in user_input.lower():
            print("Bot: I'm your chatbot assistant 🤖")

        else:
            print("Bot: You said ->", user_input)

# Run chatbot
chatbot()
```
